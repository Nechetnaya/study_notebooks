## Seaborn

```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
```

```python
import warnings; warnings.simplefilter('ignore')
```

```python
sns.set() #  стандартные параметры отображения графиков
```

```python
df = pd.read_csv('https://stepik.org/media/attachments/lesson/313078/tips.csv')
df.head()
```

```python
sns.displot(df['total_bill']) # распределение 
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind = 'reg') # гистограммы распределения двух признаков

# kind = 
# “scatter”
# “reg”
# “resid”
# “kde”
# “hex”
```

```python
# графики Плотности распределения

sns.kdeplot(df['total_bill'])

sns.rugplot(df['total_bill'])
```

```python
# отношения между всеми парами переменных
# попарное распределение количественных признаков с группировкой по какому-либо признаку "hue"

sns.pairplot(df, hue='sex', palette='Set1')
```

```python
#Построение гистограммы распределения категориального признака (количество посещений по дням недели) 
#методом countplot:

sns.countplot(x='day', data=df, palette='Set1')
```

```python
sns.countplot(y='day', data=df)
```

```python
# По умолчанию для каждой группы вычисляется среднее значение заданной количественной переменной. 
# Вместо среднего значения можно вычислить и любую другую статистическую характеристику, 
# наименование которой задается в параметре estimator.

sns.barplot(x='sex', y='total_bill', data=df)
```

```python
# Метод boxplot позволяет отобразить распределение заданного количественного признака по категориям
sns.boxplot(x="day", y="total_bill", data=df, palette='Set1')
```

```python
# при добавлении параметра hue
sns.boxplot(x='day', y='tip', data=df, hue='smoker', palette='Set1')
```

```python
sns.violinplot(x="day", y="total_bill", data=df, palette='Set2', hue='smoker')
```

```python
# stripplot рисует диаграмму рассеяния, состоящую из одной категориальной переменной

sns.stripplot(x="day", y="total_bill", data=df, palette="Set2")
```

```python
# swarmplot представляет собой так же диаграмму рассеяния, 
#точки не накладываются друг на друга

sns.swarmplot(x="day", y="total_bill",  data=df)
```

```python
# тепловые карты
df['sex_f'] = np.where(df['sex'] == 'Male', 1, 0)

cor_df = df.corr(numeric_only=True) # коэффициенты корреляции
sns.heatmap(cor_df, annot=True, cmap='coolwarm')
```

```python
# Параметры графика: размеры, подписи осей, метки на осях и прочее задаются методами matplotlib

fig, ax = plt.subplots(figsize=(8, 6))

sns.regplot(x='total_bill', y='tip', data=df, ax=ax)

ax.set_title('График распределения двух признаков')
plt.xlabel('Сумма чека')
plt.ylabel('Чаевые')

plt.show()
```

```python
sns.countplot(x='size', hue='sex', data=df)
```

## Plotly
https://plotly.com/graphing-libraries/

```python
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
x = np.arange(0, 5, 0.1)
def f(x):
    return x**2

fig = px.scatter(x=x, y=f(x))
fig.show()
```

```python
fig = go.Figure()
# Здесь будет код
# fig.add_trace(ТУТ_ТО_ЧТО_ХОТИМ_ПЕРЕДАТЬ_ДЛЯ_ОТОБРАЖЕНИЯ_И_ГДЕ)
fig.add_trace(go.Scatter(x=x, y=f(x), name='$$f(x)=x^2$$'))
fig.add_trace(go.Scatter(x=x, y=x, name='$$g(x)=x$$'))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=f(x), mode='lines+markers', name='f(x)=x<sup>2</sup>'))
fig.add_trace(go.Scatter(x=x, y=x, mode='markers', name='g(x)=x'))

fig.update_layout(legend_orientation="h")
fig.update_layout(margin=dict(l=10, r=20, t=40, b=10))
fig.update_layout(
    title="Plot Title",
    xaxis_title="x Axis Title",
    yaxis_title="y Axis Title",
)

fig.update_traces(hoverinfo="all", hovertemplate="x: %{x}<br> func: %{y}")

fig.show()

# задать свои значения отступам с помощью update_layout, у которого есть атрибут margin, 
# принимающий словарь из отступов:
# l — отступ слева
# r — отступ справа
# t — отступ сверху
# b — отступ снизу
```
