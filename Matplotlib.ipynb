## Matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
plt.figure(figsize=(8, 6), dpi=120) # настройки фигуры (размер, разрешение)

x = np.arange(-10, 11)
y = x ** 2

#plt.plot(x, y, 'r', y, x, 'g')
plt.plot(x, y)
plt.plot(y, x)

plt.grid() # сетка
plt.title('$y = x^2$\n') # заголовок
plt.xlabel('x') # подписи осей
plt.ylabel('y')

plt.savefig('1.png') # сохранение изображения 
plt.savefig('1.svg')

plt.show()
```

### График через объект `fig`

```python
fig = plt.figure() # create object

axes = fig.add_axes([0, 0, 1, 1]) # создаем объект - оси
axes.plot(x, y) # создаем график

axes2 = fig.add_axes([0.3, 0.5, 0.4, 0.4]) # создаем вторые оси
axes2.plot(x+1, y**5)
          
plt.show()
```

### Несколько графиков через объекты `subplot`

```python
fig, axes = plt.subplots() # создаем объекты фигуры и осей одновременно

axes.plot(x, y)
```

```python
fig, axes = plt.subplots(2) # создаем объекты фигуры и двоей осей (список)

axes[0].plot(x, y) # обращение через позицию в списке
axes[1].plot(y, x, 'g')
```

```python
fig, axes = plt.subplots(2, 2) # создаем объекты фигуры и оси (строки, колонки)

axes[0, 0].plot(x, y) # обращение через позицию в матрице
axes[0, 1].plot(y, x)
axes[1, 0].plot(y%2, x)
axes[1, 1].plot(y/(x+11), x**3)

plt.style.use('ggplot')
```

```python
fig, axes = plt.subplots(2, 2, figsize = (8, 8)) # задаем размер

axes[0, 0].plot(x, y)
axes[0, 0].set_xlabel('x') # подписи осей через метод
axes[0, 0].set_ylabel('y')
axes[0, 0].set_title('$y = x^2$\n')

axes[0, 1].plot(y, x, 'g')
axes[0, 1].set_xlabel('y') # подписи осей через метод
axes[0, 1].set_ylabel('x')
axes[0, 1].set_title('$x = y^2$\n')

axes[1, 0].plot(y%2, x, 'b')
axes[1, 0].set_xlabel('x') # подписи осей через метод
axes[1, 0].set_ylabel('y%2')
axes[1, 0].set_title('$y = x^2$\n')

axes[1, 1].plot(y, x**3, 'black')
axes[1, 1].set_xlabel('$x^3$') # подписи осей через метод
axes[1, 1].set_ylabel('y')
axes[1, 1].set_title('$y = x^2$\n')

fig.tight_layout() # исправляет наложение подписей

plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))

x = np.linspace(-10, 10, 100)
y = np.sin(x)

fig.add_subplot(2, 2, 1)
plt.plot(x, y)

fig.add_subplot(2, 2, 2)
plt.plot(x, y, 'b')

fig.add_subplot(2, 1, 2)
plt.plot(x, y, 'g')

plt.show()
```

```python
a = np.arange(0.1, 10, 0.1)

def f(a):
    return a ** 2
    
def g(a):
    return 1/a

def h(a):
    return a*np.sin(a)
    
```

### Color and style

`'b--' 'g-.' 'ro'` - color and stile  
`alpha=0.5` прозрачность  
`color='green'`  
`linestyle='--'`  
`linewidth=0.8`  
`marker='o'`  
`markersize=3`  
`markerfacecolor='darkgreen'`  
`linewidth=0` оставить только маркеры


```python
fig, axes = plt.subplots(2, figsize=(10, 10))

axes[0].plot(a, f(a), a, g(a))
axes[0].plot(a, h(a))

axes[0].legend(['f(x)', 'g(x)', 'h(x)']) # добавляем легенду

# легенда через label=''

axes[1].plot(a, f(a), 'b--', label='$x^2$') # LaTEX 
axes[1].plot(a, g(a), 'g-.', label=r'$\frac{1}{x}$')
axes[1].plot(a, h(a), 'ro', alpha=0.5,label=r'$x \cdot sin(x)$')
# 'b--' 'g-.' 'ro' - color and stile
# alpha=0.5 прозрачность

axes[1].set_title('functions\n', fontsize=16) #font size
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel('y', fontsize=12)
axes[1].legend(fontsize=10) #font size
fig.tight_layout()

plt.show()
```

```python
fig, axes = plt.subplots(figsize=(10, 6))

axes.plot(a, h(a), 
         color='green', 
         #linestyle='--', 
         linewidth=0, 
         marker='o',
         markersize=3,
         markerfacecolor='darkgreen')
plt.show()
```

### Установка шкалы   
`set_xlim`, `set_ylim`  
`set_yscale`, `set_xscale`  
`set_xticks'`, `set_yticks'`  
`set_xticklabels`, `set_yticklabels`

```python
fig, axes = plt.subplots(1, 2, figsize = (8, 6))

axes[0].plot(x, y)
axes[0].set_yscale('log')
axes[0].set_xticklabels(['a', 'b', 'c', 'd', 'e', 'f'])

axes[1].plot(x, y)
axes[1].set_xlim(-1, 5)
axes[1].set_ylim(-1, 20)
axes[1].set_xticks([-2, 0, 2, 4, 6])
axes[1].set_yticks(range(-2, 22, 2))

fig.tight_layout()
```
