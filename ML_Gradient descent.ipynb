## Градиентный спуск

$ X_2 = X_1 - a * f(X_1)' $  
`a` - learning rate   
`f(x)` - функция ошибки  
`X_1` - старт в случайной точке

Критерии остановки:
- значение производной достаточно маленькое 
- маленький шаг аргумента $ |X_n - X_{n-1}| $
- маленький шаг изменения функции $ |f(X_n) - f(X_{n-1})| $

### Функция нескольких переменных

Градиент функции - вектор из первых производных функции, показывает направление роста
Антиградиент - направление убывания
Перпендикуляр к касательной к функции в выбранной точке

Критерии остановки:
- маленькая длина градиента 
- маленький шаг (начальные координаты - конечные координаты -> длина)
- маленький шаг изменения функции 

### Градиентный спуск

функционал качества $ Q = \frac{1}{m}\sum_{i}^{m}{(f(x_i) - y_i)^2)} = \frac{1}{m}\sum_{i}^{m}{(w_1x_1^i +w_2x_2^i +...+w_jx_j^i - y^i)^2} $

$ \frac{dQ}{dw_1} = \frac{2}{m}\sum_{i}^{m}{(w_1x_1^i +w_2x_2^i +...+w_jx_j^i - y^i)}x_1^i $ 

...

$ \frac{dQ}{dw_j} = \frac{2}{m}\sum_{i}^{m}{(w_1x_1^i +w_2x_2^i +...+w_jx_j^i - y^i)}x_j^i $

$ Q^{vec} = (\frac{dQ}{dw_1} ... \frac{dQ}{dw_j})  $

$ w^{vec} = w^{vec} - a * Q^{vec} $

$ b = w_0, x_0^i = 1 $

стартовая точка, threshold & learning rate - подбор экспериментами

### Градиентный спуск функции одной переменной

```python
def func(x):
    return x**3 - 6 * x**2 - 15*x + 10
```

```python
x = np.linspace(-5, 10, 150)
y = func(x)
```

```python
def func_derivative(x):
    return 3 * x**2 - 12 * x - 15
```

```python
def descent_and_plot(X_start=10, learning_rate=0.01, thershold=0.0001):
    
    X_next = X_start - learning_rate * func_derivative(X_start)
    
    fig = plt.figure(figsize=(8, 6))
    
    plt.plot(x, func(x), 
             color='grey', 
             alpha = 0.8)
    
    plt.scatter([X_start, X_next], 
                [func(X_start), func(X_next)], 
                color = 'r')
    
    plt.plot([X_start, X_next], 
             [func(X_start), func(X_next)], 
             color = 'r', 
             alpha = 0.5)   
    
    plt.ylabel('$ f(x) $')
    plt.xlabel('$ x $')

    
    while abs(X_start - X_next) > thershold:
        X_start = X_next
        X_next = X_start - learning_rate * func_derivative(X_start)
        plt.scatter([X_start, X_next], [func(X_start), func(X_next)], color = 'r')
        plt.plot([X_start, X_next], [func(X_start), func(X_next)], color = 'r', alpha = 0.5) 
    
    print(X_next)

        
        
```

```python
descent_and_plot()
```

```python
descent_and_plot(X_start=10, learning_rate=0.01, thershold=0.001)
```
